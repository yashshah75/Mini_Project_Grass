# Product Profit & Category Checker

```python
p_category = input("Enter Product Category: ")
p_name = input("Enter The Product Name: ")
p_qty = int(input("Enter qty: "))
p_cost = float(input("Enter The Cost Price: "))
p_sell_price = float(input("Enter The Selling Price: "))

profit_per_unit = p_sell_price - p_cost
total_profit = profit_per_unit * p_qty
total_revenue = p_sell_price * p_qty
total_cost = p_cost * p_qty
profit_percent = (total_profit / total_cost) * 100

print("=============================== PRODUCT SUMMARY =============================== ")
print("Product Category:", p_category)
print("Product Name:", p_name)
print("Cost Price:", p_cost)
print("Selling Price:", p_sell_price)
print("Total Profit:", total_profit)
print("Total Revenue:", total_revenue)
print("Profit Percentage:", profit_percent, "%")
```
