```python
#### Vehicle Rental System using Inheritance

class Vehicle:
    def __init__(self, name, rent_per_day):
        self.name = name
        self.rent_per_day = rent_per_day

    def calculate_rent(self, days):
        return self.rent_per_day * days


class Car(Vehicle):
    def calculate_rent(self, days):
        return super().calculate_rent(days) + 100  # service charge


class Bike(Vehicle):
    def calculate_rent(self, days):
        return super().calculate_rent(days)


class Truck(Vehicle):
    def calculate_rent(self, days):
        return super().calculate_rent(days) + 500  # heavy charge


# -------- MENU SYSTEM --------
def main():
    vehicles = {
        '1': Car("Car", 1000),
        '2': Bike("Bike", 300),
        '3': Truck("Truck", 2000)
    }

    while True:
        print("\n--- Vehicle Rental System ---")
        print("1. Car (₹1000/day + ₹100(Service Charge))")
        print("2. Bike (₹300/day)")
        print("3. Truck (₹2000/day + ₹500 (Service Charge))")
        print("4. Exit")

        choice = input("Select vehicle: ")

        if choice == '4':
            print("Exiting...")
            break

        if choice in vehicles:
            try:
                days = int(input("Enter number of days: "))
                vehicle = vehicles[choice]
                rent = vehicle.calculate_rent(days)

                print(f"\nVehicle: {vehicle.name}")
                print(f"Total Rent: ₹{rent}")

            except ValueError:
                print("Invalid input! Enter valid number.")

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```
