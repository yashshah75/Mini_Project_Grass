```python
# 1. Input-Based Data Cleaner using split() and append()
data = input("Enter messy data (with spaces): ")

# split data into parts
items = data.split()

cleaned_list = []

# clean each item
for item in items:
    cleaned_item = item.strip() 
    cleaned_list.append(cleaned_item)

# output
print("Cleaned Data:", cleaned_list)
```

```python
# 20. Word Frequency Counter from File using functions

def word_frequency():
    filename = input("Enter file name: ")

    file = open(filename, "r")
    text = file.read().upper()
    
    words = text.split()
    freq = {}

    for word in words:
        if word in freq:
            freq[word] = freq[word] + 1
        else:
            freq[word] = 1

    file.close()

    print("Word Frequency:")
    for word in freq:
        print(word, ":", freq[word])

word_frequency()
```
